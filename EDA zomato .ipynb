```python
df_country=pd.read_excel("C:\\CSV Files\\Zomatodataset\\Country-Code.xlsx")
df_country.head()
```

```python
# combine these above two dataframes   pd.merge()     joining two data frammes
```

```python
final_df=pd.merge(df,df_country, on='Country Code', how='left')
```

```python
final_df.head(2)
```

```python
# to check data types
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names=final_df.Country.value_counts().index
country_names
```

```python
country_value=final_df.Country.value_counts().values
country_value
```

```python
# now plot pie chart on these info
```

```python
plt.pie(country_value[0:3],labels=country_names[0:3], autopct='%1.2f%%')
plt.show()
```

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text'])
```

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

#Observations from above rating data like, when rating between 4.5 to 4.9  that indicates Excellent.
#when rating between 4.0 to 4.4  that indicates Very Good.   and so on

```python
#plot this now
import matplotlib
```

```python
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating", y="Rating Count",data=ratings)
plt.show()
```

```python
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating", y="Rating Count",hue='Rating color',data=ratings,palette=['Black','red','orange','green','yellow','green','dark green'])
plt.show()
```

```python
sns.countplot(x='Rating color',data=ratings,palette=['Black','red','orange','green','yellow','green','dark green'])
```
