```python
plt.plot(x,y1,color='g',linestyle=':',linewidth=4)
plt.plot(x,y2,color='b',linestyle='-',linewidth=3)
plt.title("Line Plot")
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.grid(True)
plt.show
```
