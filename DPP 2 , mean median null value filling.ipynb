```python
missing_value_per=df.isnull().sum()/df.shape[0]*100
missing_value_per
```

```python
missing_value_gre_20=missing_value_per[missing_value_per>20].keys()
missing_value_gre_20
```

```python
df2=df.drop(columns=missing_value_gre_20)
df2.shape
```

```python
df3_num=df2.select_dtypes(include=['int64','float64'])
df3_num.head()
```

```python
sns.heatmap(df3_num.isnull())
```

```python
df4=df3_num[df3_num.isnull().any(axis=1)]
df4
```

```python
df4.isnull().sum()
```

```python
missing_num_var=[var for var in df3_num.columns if df3_num[var].isnull().sum() >0 ]
missing_num_var
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var) :
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var],bins=20, kde_kws={'linewidth':2,'color':'#DC143C'})
```

```python
df4_num_mean=df3_num.fillna(df3_num.mean())
```

```python
df4_num_mean.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var],bins=20,kde_kws={'linewidth':8,'color':'red'},label="origional")
    sns.distplot(df4_num_mean[var],bins=20,kde_kws={'linewidth':8,'color':'black'},label="cleaned")
    plt.legend()
```

```python
df5_num_median=df3_num.fillna(df3_num.median())
df5_num_median.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var],bins=20,hist=False,kde_kws={'linewidth':8,'color':'red'},label="origional")
    sns.distplot(df4_num_mean[var],bins=20,hist=False,kde_kws={'linewidth':5,'color':'black'},label="mean")
    sns.distplot(df5_num_median[var],bins=20,hist=False,kde_kws={'linewidth':3,'color':'blue'},label="median")
    plt.legend()
```
